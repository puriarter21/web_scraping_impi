```python
import json
import pandas as pd

with open("4.marcia_details.jsonl", encoding="utf-8") as f:
    data = [json.loads(line) for line in f if line.strip()]

rows = []
for item in data:
    record_id = item.get("id")
    history = (
        item.get("details", {})
            .get("historyData", {})
            .get("historyRecords", [])
    )
    
    descriptions = {r.get("description", "").strip() for r in history if r.get("description")}
    
    rows.append({"id": record_id, "_descriptions": descriptions})

# Todas las descripciones únicas en el dataset
all_descriptions = sorted({d for row in rows for d in row["_descriptions"]})
print(f"Trámites únicos encontrados: {len(all_descriptions)}")
for d in all_descriptions:
    print(f"  {d}")

# Construir df con una columna binaria por trámite
records = []
for row in rows:
    entry = {"id": row["id"]}
    for desc in all_descriptions:
        entry[desc] = 1 if desc in row["_descriptions"] else 0
    records.append(entry)

df_tramites = pd.DataFrame(records)


rows = []
for item in data:
    record_id = item.get("id")
    history = item.get("historyData", {}).get("historyRecords", [])   # ← path correcto
    
    descriptions = {r.get("description", "").strip() for r in history if r.get("description")}
    rows.append({"id": record_id, "_descriptions": descriptions})

all_descriptions = sorted({d for row in rows for d in row["_descriptions"]})
print(f"Trámites únicos encontrados: {len(all_descriptions)}")
for d in all_descriptions:
    print(f"  {d}")

records = []
for row in rows:
    entry = {"id": row["id"]}
    for desc in all_descriptions:
        entry[desc] = 1 if desc in row["_descriptions"] else 0
    records.append(entry)

df_tramites = pd.DataFrame(records)
```
