```python
import json
import pandas as pd

with open("4.marcia_details.jsonl", encoding="utf-8") as f:
    data = [json.loads(line) for line in f if line.strip()]

def aplanar_registro(r: dict) -> dict:
    fila = {}

    # ── id ──────────────────────────────────────────────────────
    fila["id"] = r.get("id", "")

    # ── generalInformation ──────────────────────────────────────
    gen = r.get("details", {}).get("generalInformation", {})
    fila["title"]              = gen.get("title", "")
    fila["applicationNumber"]  = gen.get("applicationNumber", "")
    fila["registrationNumber"] = gen.get("registrationNumber", "")
    fila["applicationDate"]    = gen.get("applicationDate", "")
    fila["publicationDt"]      = gen.get("publicationDt", "")
    fila["registrationDate"]   = gen.get("registrationDate", "")
    fila["expiryDate"]         = gen.get("expiryDate", "")
    fila["appType"]            = gen.get("appType", "")

    # ── productsAndServices → solo classes ──────────────────────
    prods = r.get("details", {}).get("productsAndServices", [])
    fila["classes"] = ", ".join(str(p.get("classes", "")) for p in prods if p.get("classes") != "")

    # ── viennaCodes ─────────────────────────────────────────────
    fila["viennaCodes"] = r.get("details", {}).get("trademark", {}).get("viennaCodes", "")

    # ── ownerInformation ────────────────────────────────────────
    owners = r.get("details", {}).get("ownerInformation", {}).get("owners", [])
    owner  = owners[0] if owners else {}
    fila["Addr"] = "; ".join(owner.get("Addr", []))
    fila["Cry"]  = "; ".join(owner.get("Cry",  []))
    fila["Name"] = "; ".join(owner.get("Name", []))

    # ── historyData → solo primer registro ──────────────────────
    records = r.get("historyData", {}).get("historyRecords", [])
    hist    = records[0] if records else {}
    #fila["description"]     = hist.get("description",     "")
    fila["email"]           = hist.get("email",           "")
    fila["dateOfConclusion"] = hist.get("dateOfConclusion", "")
    fila["receptionYear"]   = hist.get("receptionYear",   "")
    fila["startDate"]       = hist.get("startDate",       "")

    # ── pdf_links ────────────────────────────────────────────────
    fila["pdf_links"] = "; ".join(r.get("pdf_links", []))

    return fila

filas = [aplanar_registro(r) for r in data]
df    = pd.DataFrame(filas)
print("------------------------------------")

print(df.shape)
print(df.columns.tolist())

df_mex = df[df["Cry"].str.contains("MEXICO", na=False)]


print("------------------------------------")
print(df_mex.shape)

print("------------------------------------")
```

```python
import re
# Addrs cleaning
df_address = df_mex[["id", "Addr"]].copy()

def limpiar_direccion(addr):
    if not isinstance(addr, str):
        return addr
    texto = addr.lower()
    texto = re.sub(r'\bnum\.?\b', '', texto, flags=re.IGNORECASE)
    texto = re.sub(r'\bext\.?\b', '', texto, flags=re.IGNORECASE)
    texto = re.sub(r'[\s,.\(\)\[\]]+', ' ', texto)
    texto = texto.strip()
    return texto

# Aplicar la limpieza a la columna
df_address["Addr"] = df_address["Addr"].apply(limpiar_direccion)
```
